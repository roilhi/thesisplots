```python
import os, sys 
import numpy as np
from os import listdir, path
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
```

```python
#Reading the paths 
PcgPath = '/Users/roilhiibarra/Dropbox/TesisDocVarios/PhysionetChallenge/PhysionetSoundsDatabase/validation'
signalName = 'a0007.wav'
x_t = path.join(PcgPath,signalName)
on = 39200
off = 40460
```

```python
fs, signal = wav.read(x_t)
signal = signal[on:off]
signal = signal/max(signal)
font = {'family' : 'sans-serif',
        'weight' : 'bold',
        'size'   : 14}

plt.rc('font', **font)
plt.title('Cardiac cycle of a normal PCG')
plt.plot(signal)
plt.ylabel('Amplitude')
plt.xlabel('Time (samples)')
plt.annotate('S1',xy=(300,0.5),xytext=(100,0.50))
plt.annotate('S2',xy=(900,0.6),xytext=(700,0.50))
xcoords = [200, 400,800,1060]

for xc in xcoords:
    plt.axvline(x=xc,c='r')

plt.grid()
plt.show()
```
